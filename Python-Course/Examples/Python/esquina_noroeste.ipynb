### Esquina Noroeste

De parte del grupo 1, realizamos el método de esquina noroeste para la siguiente tabla de demanda y oferta:





```python

import numpy as np
import matplotlib.pyplot as plt
import networkx as nx
# Definir datos
Etiquetas = {"Destinos": ['Managua','Ticuantepe', 'San Juan Del Sur', 'Juigalpa'],
             "Origenes": ['Masaya', "Granada", "Rivas", "Estelí", "Matagalpa" ],
            }
costos =np.array([[100, 90, 150, 200],
          [180, 110, 250, 300],
          [320, 150, 180, 350],
          [350, 300, 340, 380],
          [400, 400, 350, 400]
          ])

oferta = np.array([[1200],
          [1600],
          [2500],
          [3200],
          [3000],
         ])

demanda = np.array([[3000, 4000, 2500, 2000]])
asignaciones = np.zeros((len(Etiquetas["Origenes"]), len(Etiquetas["Destinos"])))


def esquina_noroeste(costos, oferta, demanda):
    asignaciones = np.zeros((oferta.size, demanda.size))    
    i, j = 0, 0
    oferta = np.squeeze(oferta)
    demanda = np.squeeze(demanda)
    costo_total = 0
    while i < costos.shape[0] and j < costos.shape[1]:
        if np.all(oferta[i]) == 0 or np.all(demanda[j]) == 0:
            if oferta[i] == 0:
                i += 1
            else:
                j += 1
        else:
            cantidad = min(oferta[i], demanda[j])
            asignaciones[i, j] = cantidad
            costo_total += cantidad * costos[i, j]
            oferta[i] -= cantidad
            demanda[j] -= cantidad
            if oferta[i] == 0:
                i += 1
            if demanda[j] == 0:
                j += 1  
            
    return asignaciones, costo_total

asignaciones, costo_total = esquina_noroeste(costos, oferta, demanda)
print(asignaciones)
print(f'El costo total es {costo_total}')
for i in range(len(Etiquetas["Origenes"])):
        for j in range(len(Etiquetas["Destinos"])):
            plt.annotate(str(asignaciones[i, j]), xy=(j, i), ha="center", va="center")
plt.imshow(asignaciones, cmap='Greens')
plt.colorbar()
plt.xticks(range(len(Etiquetas['Destinos'])), Etiquetas['Destinos'], rotation=90)
plt.yticks(range(len(Etiquetas['Origenes'])), Etiquetas['Origenes'])
plt.title('Tabla de asignaciones')
plt.show()      
```

```python
G = nx.DiGraph()
for i in range(len(Etiquetas["Origenes"])):
    G.add_node(Etiquetas["Origenes"][i], pos=(0,i))

# Agregar nodos de destino
for i in range(len(Etiquetas["Destinos"])):
    G.add_node(Etiquetas["Destinos"][i], pos=(2,i))

# Agregar arcos con etiquetas de costo y asignaciones
for i in range(len(Etiquetas["Origenes"])):
    for j in range(len(Etiquetas["Destinos"])):
        if asignaciones[i][j] > 0:
            G.add_edge(Etiquetas["Origenes"][i], Etiquetas["Destinos"][j], weight=costos[i][j], label=str(asignaciones[i][j]))

# Obtener posiciones de los nodos
pos = nx.get_node_attributes(G, 'pos')

# Obtener etiquetas de arcos
labels = {(i, j): G[i][j]['label'] for i, j in G.edges()}

# Dibujar grafo
nx.draw(G, pos, with_labels=True, font_size=8, node_size=3500, node_color='white', edgecolors='black')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels, font_size=10)

# Mostrar grafo
plt.axis('off')
plt.show()
```
